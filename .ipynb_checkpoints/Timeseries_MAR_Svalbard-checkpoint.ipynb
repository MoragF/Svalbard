```python
import xarray as xr
import numpy as np
def MAR(x):
    ds= xr.open_mfdataset('/exports/csce/datastore/geos/groups/geos_EO/Databases/MAR/Svalbard-RA/MARv3.11.*.nc')
    timeseries = ds[x].groupby('TIME.day').mean(dim=('Y20_218','X20_340'))  
    return timeseries.plot()
    
```

```python
timeseries.plot()
```
