```python
from affine import Affine
tif = xr.open_rasterio('07_rgi60_Svalbard_Austfonna_corr_land_100m.tif')

import rasterio
from rasterio.features import shapes
mask = None
with rasterio.Env():
    with rasterio.open('07_rgi60_Svalbard_Austfonna_corr_land_100m.tif') as src:
        image = src.read(1) # first band
        results = (
        {'properties': {'raster_val': v}, 'geometry': s}
        for i, (s, v) 
        in enumerate(
            shapes(image, mask=mask, transform=src.transform)))
```

```python
#ncfile = netcdf('MARv3.11.2-6km-daily-ERA5-2000.nc')
#ncfile.variables
ds= xr.open_dataset('MARv3.11.2-6km-daily-ERA5-2000.nc')
#Fixing coordinates to lat and lon in prep for attempting to use mask
ds2=ds.set_coords(['LAT','LON'])
ds2.coords['y'] = tif.y.values
ds2.coords['x'] = tif.x.values
ds2.coords['band'] = tif.band.values
ds2.coords['mask']= (('band','y','x'),tif)
ds2.coords['mask2']= (('band','y','x'),tif.where(tif==1))
ds2
```

```python

nx=[]
ny=[]
lon=[]
lat=[]
diff=0.03
g=min(ds2.y)-0.1
h=max(ds2.y)+0.1
k=min(ds2.x)-0.1
l=max(ds2.x)+0.1
for i in range(len(ds2.Y20_218.values)):
    for j in range(len(ds2.X20_340.values)):
        a=ds2.LAT[i][j].values
        b=ds2.LON[i][j].values
        if g<=a<=h and k<=b<=l:
            e=min(ds2.y.values, key=lambda n:abs(n-a))
            f=min(ds2.x.values, key=lambda n:abs(n-b))
            lat.append(a)
            lon.append(b)
            if abs(e-a)<diff and abs(f-b)<diff:
                c=list(ds2.y.values).index(e)
                d=list(ds2.x.values).index(f)
                if ds2.mask[0][c][d]==1:
                    ny.append(ds2.Y20_218[i].values)
                    nx.append(ds2.X20_340[j].values)
```
