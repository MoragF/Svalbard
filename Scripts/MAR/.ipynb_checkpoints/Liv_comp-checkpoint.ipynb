```python
def multi_ELE_test(typ,ntype,title,npng):
    fig, axes = plt.subplots(3,2,figsize=(18,14), sharex=True)
    
    for i in range(3):
        for j in range(2):
            ds= xr.open_dataset('/exports/csce/datastore/geos/groups/geos_EO/Databases/MAR/Svalbard-RA/Svalbard_Masked/Region_Tmsr/Svalbard_'+str(typ)+str((1*(i+1))+i+j)+'.nc')
            ds['time']= pd.date_range('2010-01-01','2020-12-31')
            SMB= ds.smb_mean

            matdata=io.loadmat('/geos/d14/CS2/dhdt/Timeseries/Svalbard_uncorr_MF/Svalbard_'+str(typ)+str((1*(i+1))+i+j)+'_MF/tmsr_rgi60_Svalbard_'+str(ntype)+'_R'+str((1*(i+1))+i+j)+'.mat')
            elevation = matdata['Tmsr'][0]/0.00085
            t = matdata['time'][0]
            timepy= pd.to_datetime(t-719529, unit='D')
            E_ds = xr.Dataset(data_vars = dict(elev= (["t"],elevation+(0-elevation[0]))),coords=dict(t=timepy))

            labels = ['MAR SMB','Recur ','Livia']
            plt.rcParams['font.size'] = '12'
            color = 'tab:red'
            color2 = 'tab:cyan'
            color3 = 'tab:green'
            
            Liv_Elev = xr.open_dataset("Liv_Elevation.nc")
            d_l=[Liv_Elev.LR1/0.00085,Liv_Elev.LR2/0.00085,Liv_Elev.LR3/0.00085,Liv_Elev.LR4/0.00085,Liv_Elev.LR5/0.00085,Liv_Elev.LR6/0.00085]
            axes[i,j].plot(SMB.time.sel(time=slice('2010-07-16','2019-12-31')), SMB.sel(time=slice('2010-07-16','2019-12-31')).cumsum(), color=color2)
            
            axes[i,j].legend(labels)
            axes[i, j].set_title('Region '+str((1*(i+1))+i+j))
            axes[i,j].plot(E_ds.t, E_ds.elev, color=color)
            axes[i,j].plot(pd.to_datetime(Liv_Elev.t), (d_l[(1*(i+1))+i+j-1]+(0-d_l[(1*(i+1))+i+j-1][0])), color=color3)
            axes[i,j].axhline(y=0,linestyle='dashed',color='gray')
            axes[i,j].legend(labels)
            
    for ax in axes.flat:
        ax.set(xlabel='Time', ylabel=str(SMB.units))
    for ax in axes.flat:
        ax.label_outer()
    fig.suptitle('Svalbard '+str(title)+' Terminating')
    #fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.savefig(npng)
    plt.show()
```

```python
def MARplot(matfile,ncfile,title,npng,livia):
    ds= xr.open_dataset(ncfile)
    ds['time']= pd.date_range('2010-01-01','2020-12-31')
    SMB= ds.smb_mean
    TSH=ds.tsh_mean
    SMB_diff = SMB - SMB.mean('time')
    TSH_diff = TSH - TSH.mean('time')

    matdata=io.loadmat(matfile)
    elevation = matdata['Tmsr'][0] /0.00085
    t = matdata['time'][0]
    timepy= pd.to_datetime(t-719529, unit='D')
    E_ds = xr.Dataset(data_vars = dict(elev= (["t"],elevation+(0-elevation[0]))),coords=dict(t=timepy))
    diff= E_ds.elev-E_ds.elev.mean('t')
    
    Liv_Elev= pd.read_csv(livia, parse_dates=[1], index_col=2)
    Liv = xr.Dataset(data_vars = dict(L_elev= (["t"],Liv_Elev.changes.values)),coords=dict(t=Liv_Elev.dates_datetime.values))

    fig, (ax1,ax3,ax4) = plt.subplots(3,figsize=(16,15), sharex=True)
    plt.suptitle(title, fontsize=16)

    color1 = 'tab:red'
    color2 = 'tab:blue'
    ax1.set_ylabel('mmWE/day', color=color1, fontsize=15)
    a,= ax1.plot(E_ds.t, E_ds.elev, color=color1,label = 'Recur dh/dt')
    c,= ax1.plot(Liv.t, Liv.L_elev, color=color2,label = 'Livia dh/dt')
    p = [a,c]
    ax1.legend(p, [p_.get_label() for p_ in p])
    
    color = 'tab:green'
    color2= 'tab:purple'
    ax3.set_ylabel('mmWE/day', color=color, fontsize=15)  
    ax3.plot(TSH.time.sel(time=slice('2010-07-18','2019-12-31')), TSH.sel(time=slice('2010-07-18','2019-12-31')), color=color)
    ax3.plot(SMB.time.sel(time=slice('2010-07-18','2019-12-31')), SMB.sel(time=slice('2010-07-18','2019-12-31')).cumsum(), color=color2, label = 'SMB MAR')
    ax3.plot(E_ds.t, E_ds.elev, color=color1,label = 'Recur dh/dt') 
    ax3.legend(['TSH MAR','SMB MAR'])
    
    color = 'tab:green'
    color2= 'tab:purple'
    ax4.set_ylabel('mmWE/day', color=color, fontsize=15)  
    ax4.plot(TSH.time.sel(time=slice('2010-07-18','2019-12-31')), TSH_diff.sel(time=slice('2010-07-18','2019-12-31')), color=color)
    ax4.plot(SMB.time.sel(time=slice('2010-07-18','2019-12-31')), SMB_diff.sel(time=slice('2010-07-18','2019-12-31')).cumsum(), color=color2, label = 'SMB MAR')
    ax4.plot(E_ds.t, diff, color=color1,label = 'Recur dh/dt') 
    ax4.legend(['TSH MAR diff','SMB MAR diff', 'Recur dh/dt diff'])
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # otherwise the right y-label is slightly clipped
    plt.savefig(npng)
    plt.show()
   
```
